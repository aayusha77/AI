```python
df1 = pd.read_csv('assignment-data.csv')
df2 = pd.read_csv('assignment-data2.csv')
df3 = pd.read_csv('linear-regression-data1.csv')
```

```python

X1 = torch.tensor(df1['x'].values, dtype=torch.float32)
y1 = torch.tensor(df1['y'].values, dtype=torch.float32)
```

```python
X2 = torch.tensor(df2['x'].values, dtype=torch.float32)
y2 = torch.tensor(df2['y'].values, dtype=torch.float32)
```

```python
X3 = torch.tensor(df3['x'].values, dtype=torch.float32)
y3 = torch.tensor(df3['y'].values, dtype=torch.float32)
```

```python

print(f"Data-1: {X1.shape, y1.shape}")
print(f"Data-2: {X2.shape, y2.shape}")
print(f"Data-3: {X3.shape, y3.shape}")
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)


X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)


X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
```

```python
plt.figure(figsize=(18, 6))

# Plot for Data-1
plt.subplot(1, 3, 1) # 1 row, 3 columns, 1st subplot
plt.scatter(X1_train, y1_train, label='Training Data', alpha=0.7)
plt.scatter(X1_test, y1_test, label='Testing Data', alpha=0.7)
plt.title('Data-1: Training vs. Testing Split')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Plot for Data-2
plt.subplot(1, 3, 2) # 1 row, 3 columns, 2nd subplot
plt.scatter(X2_train, y2_train, label='Training Data', alpha=0.7)
plt.scatter(X2_test, y2_test, label='Testing Data', alpha=0.7)
plt.title('Data-2: Training vs. Testing Split')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Plot for Data-3
plt.subplot(1, 3, 3) # 1 row, 3 columns, 3rd subplot
plt.scatter(X3_train, y3_train, label='Training Data', alpha=0.7)
plt.scatter(X3_test, y3_test, label='Testing Data', alpha=0.7)
plt.title('Data-3: Training vs. Testing Split')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```

**Among the three datasets, Data-3 is the only one that can be accurately represented using a linear model because its data points follow an approximately straight-line relationship. In contrast, Data-1 and Data-2 exhibit non-linear, curved patterns that resemble quadratic relationships. Since a linear equation cannot effectively capture these curves, it would not be an appropriate choice for modeling Data-1 and Data-2.**
