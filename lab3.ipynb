# Theory

#Introduction to PyTorch

PyTorch is an open-source machine learning and deep learning framework developed by Meta (Facebook). It provides a flexible platform for building, training, and deploying neural networks. PyTorch uses tensors as its primary data structure and supports automatic differentiation, making it widely used for research and industrial AI applications.

#Tensor

A tensor is the fundamental data structure in PyTorch. It is a multi-dimensional array used to store numerical data. Tensors are similar to NumPy arrays but can utilize GPUs for faster computation.

#Types of Tensors
1. Scalar

A scalar is a tensor containing a single value and has zero dimensions.

Example: tensor(7)

2. Vector

A vector is a one-dimensional tensor containing multiple values.

Example: [7, 8, 9, 10]

3. Matrix

A matrix is a two-dimensional tensor arranged in rows and columns.

Example:

[[1,2,3],
 [4,5,6],
 [7,8,9]]

4. Higher-Dimensional Tensor

A tensor with three or more dimensions is used to represent complex data such as images, videos, and datasets.

#Tensor Creation

PyTorch provides various methods for creating tensors:

torch.tensor() – Creates a tensor from data.

torch.rand() – Creates a tensor with random values.

torch.zeros() – Creates a tensor filled with zeros.

torch.ones() – Creates a tensor filled with ones.

torch.arange() – Creates a tensor with a sequence of values.

Tensor Attributes

#Important tensor properties include:

#Shape

Represents the dimensions of a tensor.

#Datatype (dtype)

Specifies the type of data stored in the tensor, such as:

float32

float16

int64

#Tensor Operations


Tensor operations are essential in machine learning and deep learning.

Arithmetic Operations

Addition

Subtraction

Multiplication

Division

Element-wise Multiplication

Each element of one tensor is multiplied by the corresponding element of another tensor.

#Matrix Multiplication

Matrix multiplication combines two matrices following linear algebra rules. For multiplication to be valid, the inner dimensions must match.

Example:

(2 × 3) × (3 × 2) = (2 × 2)

#Tensor Aggregation Operations

Aggregation operations summarize tensor values.

#Common Functions
min() – Returns the minimum value.

max() – Returns the maximum value.

mean() – Returns the average value.

sum() – Returns the sum of all elements.

argmin() – Returns the index of the minimum value.

argmax() – Returns the index of the maximum value.

Tensor Reshaping

Reshaping changes the dimensions of a tensor without altering its data.

reshape()

Changes the tensor shape while maintaining the same elements.

view()

Returns a new view of the tensor sharing the same memory.

stack()

Combines multiple tensors along a new dimension.

Squeeze and Unsqueeze Operations

squeeze()

Removes dimensions of size 1 from a tensor.

unsqueeze()

Adds a new dimension of size 1 at a specified position.

These operations help maintain compatible tensor shapes during neural network computations.

#Permutation of Dimensions

torch.permute() rearranges the order of tensor dimensions without changing the data itself. It is commonly used when working with image data where dimensions need to be reordered.

#Tensor Indexing

Indexing allows access to specific tensor elements, rows, columns, or slices.

Examples include:

Accessing individual elements

Selecting rows and columns

Slicing tensors using : operator

#Conversion Between NumPy and PyTorch

PyTorch and NumPy can easily exchange data.

#NumPy to Tensor
torch.from_numpy()
#Tensor to NumPy
tensor.numpy()

This interoperability is useful for scientific computing and data preprocessing.

GPU Support in PyTorch

PyTorch supports GPU acceleration using CUDA. Computations performed on GPUs are significantly faster than CPUs for large-scale deep learning tasks.

#Common functions:

torch.cuda.is_available()

torch.cuda.device_count()

tensor.to('cuda')

If no GPU is available, PyTorch automatically uses the CPU.

```python
# # Shapes need to be in the right way  
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)

# torch.matmul(tensor_A, tensor_B) # (this will error)
```

```python
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output) 
print(f"\nOutput shape: {output.shape}")
```

```python
print(x[:, 0]) # Get all values of 0th dimension and the 0 index of 1st dimension
print(x[:, :, 1])# Get all values of 0th & 1st dimensions but only index 1 of 2nd dimension
print(x[:, 1, 1]) # Get all values of the 0 dimension but only the 1 index value of the 1st and 2nd dimension
print(x[0, 0, :]) # same as x[0][0]
```
